```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
data['petal_length'] = iris.target
```

```python
print(data.head())
```

```python
print(data.info())
```

```python
print(data.describe())
```

```python
print(data.isnull().sum())
```

```python
sns.pairplot(data, hue='petal_length', palette='deep')
```

```python
plt.title("Pairplot of Iris Dataset by Target Class")
```

```python
plt.show()
```

```python
plt.figure(figsize=(12, 6))
```

```python
for i, feature in enumerate(iris.feature_names):
    plt.subplot(2, 2, i + 1)
```

```python
 sns.boxplot(x='petal_length', y=feature, data=data, palette='plasma')
```

```python
plt.title(f"Box Plot of {feature} by Target Class")
```

```python
plt.tight_layout()
plt.show()
```

```python
correlation_matrix = data.corr()
```

```python
plt.figure(figsize=(10, 8))
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
```

```python
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
```

```python
for i, feature in enumerate(iris.feature_names):
    plt.subplot(2, 2, i + 1)
```

```python
for target_class in iris.target_names:
        sns.histplot(data[data['petal_length'] == target_class][feature], bins=20, kde=True, label=target_class)
```

```python
plt.legend()
```

```python
plt.title(f"Histogram of {feature} by Target Class")
```

```python
plt.tight_layout()
plt.show()
```

```python
# Customized PairGrid for more control over pairplots
g = sns.PairGrid(data, hue='petal_length', palette='deep')
g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot, cmap='coolwarm')
g.add_legend()
g.fig.suptitle("Customized PairGrid of Iris Dataset by Target Class")
plt.show()
```

```python
# Countplot of target classes
plt.figure(figsize=(8, 6))
sns.countplot(x='petal_length', data=data, palette='bright')
plt.title("Count of Each Target Class")
plt.xlabel("Target Class")
plt.ylabel("Count")
plt.show()
```

```python
# Violin plots for each feature separated by target class
plt.figure(figsize=(12, 6))
for i, feature in enumerate(iris.feature_names):
    plt.subplot(2, 2, i + 1)
    sns.violinplot(x='petal_length', y=feature, data=data, palette='colorblind', inner='quartile')
    plt.title(f"Violin Plot of {feature} by Target Class")
plt.tight_layout()
plt.show()
```
